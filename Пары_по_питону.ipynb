```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import  matplotlib as mpl
import numpy as np

x=y=np.linspace(-3,3,74)
X,Y = np.meshgrid(x,y)

R = np.sqrt(X**2+Y**2)
Z = np.sin(4*R)/R
fig,ax =plt.subplots(1,3,figsize=(14,4), subplot_kw=dict(projection='3d'))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax[0].plot_surface(X,Y,Z,rstride=1,cstride=1,linewidth=0,
antialiased=False,norm=norm,cmap=mpl.cm.Blues)
cb=fig.colorbar(p,ax=ax[0],shrink=0.6) # вертикальная полоска у первого графика 

ax[0].set_xlabel("$x$", fontsize=16)
ax[0].set_ylabel("$y$", fontsize=16)
ax[0].set_zlabel("$z$", fontsize=16)

ax[1].plot_wireframe(X,Y,Z,rstride=3,cstride=3,color="darkgrey")
ax[1].set_title("plot_wireframe")


ax[2].contour(X,Y,Z, zdir='z',offset=0,norm=norm,cmap=mpl.cm.Blues)
#ax[2].contour(X,Y,Z, zdir='y',offset=0,norm=norm,cmap=mpl.cm.Blues)

ax[2].set_xlabel("$x$", fontsize=16)
ax[2].set_ylabel("$y$", fontsize=16)
ax[2].set_zlabel("$z$", fontsize=16)
plt.show()
fig.savefig("Sb0.png",dpi=100) # results int 4*160x4*120 px image
```

```python
# функция, в которой задаются x,y,z
from mpl_toolkits.mplot3d import Axes3D
def makeData():
  x=np.arange(-10,10,0.1)
  y=np.arange(-10,10,0.1)
  xgrid,ygrid = np.meshgrid(x,y)
  zgrid = np.sin(xgrid)*np.sin(ygrid) / (xgrid*ygrid)
  
  return xgrid,ygrid,zgrid

x,y,z = makeData()

fig = plt.figure()
ax= Axes3D(fig)

ax.plot_surface(x,y,z,rstride=4,cstride=4,cmap=mpl.cm.jet)

plt.show()
```

```python
# пример с mgrid вместо meshgrid
u,v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j] # j - вместо шага в 20 мы делаем кол-во точек = 20
x = np.cos(u)**np.sin(v)
y = np.sin(u)**np.sin(v)
z=np.cos(v)
fig = plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z,cmap='inferno')

plt.show()
```
